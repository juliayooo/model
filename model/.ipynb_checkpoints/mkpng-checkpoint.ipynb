```python
import glob
import os

# Assign directory
directory = "./resized_augment2_2/"

filenames = []

# Iterate over files in directory
for name in glob.glob(directory + '/*/*/*'):
    filenames += [name]
```

```python
import cv2

ct = 0
for file in filenames:
    img = cv2.imread(file)
    if 'test' in file:
        new_name = './data2/test/'
    if 'train' in file:
        new_name = './data2/train/'
    if 'val' in file:
        new_name = './data2/val/'
    if 'no' in file:
        new_name = new_name + 'no/'
    if 'yes' in file:
        new_name = new_name + 'yes/'
    new_name = new_name + 'person_' + f'{ct:05}' + '.png'

    img = cv2.imwrite(new_name, img)

    ct += 1
```
